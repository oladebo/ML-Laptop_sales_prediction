- #### D. Encoding the categorical columns to numerical

```python
# Find the Unique Value

df['Brand'].unique()
```

```python
df['Brand'] = df['Brand'].map({'Asus':0,'Acer':1,'Lenovo':2,'HP':3,'Dell':4})
```

```python
df['Brand'].unique()
```

```python
df.head()
```

```python
df['Price'] = df['Price'].round().astype(int)
```

```python
df.head()
```

```python
df.info()
```

- ### E. Divide our dataset into X input feature and y to Target feature

```python
df.head(3)
```

```python
X = df.drop(columns= ['Price'])
y = df['Price']
```

```python
X.head()
```

- ### F. Spliting dataset into Training set/Testing set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.20, random_state=42)
```

```python
# Feature scalling

from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

- #### G. Building Model for Training

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)

svm = SVR()
svm.fit(X_train, y_train)

rf = RandomForestRegressor()
rf.fit(X_train, y_train)

gbr = GradientBoostingRegressor()
gbr.fit(X_train, y_train)

dt = DecisionTreeRegressor()
dt.fit(X_train, y_train)
```

- #### H. Prediction on Test Data

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_pred1 = lr.predict(X_test)
```

```python
svm = SVR()
svm.fit(X_train, y_train)
```

```python
y_pred2 = svm.predict(X_test)
```

```python
rf = RandomForestRegressor()
rf.fit(X_train, y_train)
```

```python
y_pred3 = rf.predict(X_test)
```

```python
gbr = GradientBoostingRegressor()
gbr.fit(X_train, y_train)
```

```python
y_pred4 = gbr.predict(X_test)
```

```python
dt = DecisionTreeRegressor()
dt.fit(X_train, y_train)
```

```python
y_pred5 = dt.predict(X_test)
```

- #### I. Evaluating the Algorithm

```python
# That is Perfomance of our Model

from sklearn import metrics
```

```python
score1 = metrics.r2_score(y_test, y_pred1)
score2 = metrics.r2_score(y_test, y_pred2)
score3 = metrics.r2_score(y_test, y_pred3)
score4 = metrics.r2_score(y_test, y_pred4)
score5 = metrics.r2_score(y_test, y_pred5)
```

```python
print(score1,score2,score3,score4,score5)
```

```python
final_data = pd.DataFrame({'Models':['LR','SVR','RF','GBR','DT'],
              'R2_SCORE':[score1,score2,score3,score4,score5]})
```
